```python
# Function to determine type and gender for nouns
def determine_noun_gender(word):
    if word.startswith("der "):
        return "Masculine"
    elif word.startswith("die "):
        return "Feminine"
    elif word.startswith("das "):
        return "Neuter"
    else:
        return None

# Function to determine type of word
def determine_type(word):
    if "," in word and re.match(r".*, .*", word):
        return "verb"
    elif word.startswith(("der ", "die ", "das ")):
        return "noun"
    elif re.match(r"^\w+$", word):  # Single word entry
        return "adjective"
    elif re.match(r"^\(.*\)$", word):  # Entry in parentheses
        return "example"
    else:
        return "adverb"

# Function to determine type of adverb
def determine_adverb_type(word):
    for adverb in local_adverbs:
        if adverb in word:
            return "local"
    for adverb in causal_adverbs:
        if adverb in word:
            return "causal"
    for adverb in temporal_adverbs:
        if adverb in word:
            return "temporal"
    for adverb in modal_adverbs:
        if adverb in word:
            return "modal"
    return "unknown"

# Clean and process the raw data
cleaned_data = []
current_word = None
current_conjugation = None
current_example = None

for entry in raw_data:
    entry = entry.strip()

    # Skip irrelevant entries
    if any(keyword in entry.lower() for keyword in ["kapitel", "aspekte", "seite", "modul"]):
        continue

    word_type = determine_type(entry)

    if word_type == "verb":
        current_word = entry.split(",")[0].strip()
        current_conjugation = entry
        cleaned_data.append({
            "word": current_word,
            "type": word_type,
            "gender": None,
            "conjugation": current_conjugation,
            "example_sentence": None
        })
    elif word_type == "noun":
        gender = determine_noun_gender(entry)
        current_word = entry.split(",")[0].strip()
        cleaned_data.append({
            "word": current_word,
            "type": word_type,
            "gender": gender,
            "conjugation": None,
            "example_sentence": None
        })
    elif word_type == "adjective":
        cleaned_data.append({
            "word": entry,
            "type": word_type,
            "gender": None,
            "conjugation": None,
            "example_sentence": None
        })
    elif word_type == "adverb":
        adverb_type = determine_adverb_type(entry)
        cleaned_data.append({
            "word": entry,
            "type": word_type,
            "gender": None,
            "conjugation": None,
            "example_sentence": None,
            "adverb_type": adverb_type
        })
    elif word_type == "example":
        if cleaned_data and cleaned_data[-1]["example_sentence"] is None:
            cleaned_data[-1]["example_sentence"] = entry.strip("()")
```
